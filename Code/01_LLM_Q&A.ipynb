```python
import os
from langchain_openai import OpenAI
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from google.colab import userdata

# Set up your OpenAI API key
os.environ['OPENAI_API_KEY'] = userdata.get('OPENAI_API_KEY')

# Define a prompt template
prompt_template = PromptTemplate(
    input_variables=["user_input"],
    template="You are a helpful chatbot. User: {user_input} Chatbot:"
)

# Initialize the OpenAI LLM
llm = OpenAI()

# Create the LLM Chain, use 'prompt' instead of 'prompt_template'
chat_chain = LLMChain(llm=llm, prompt=prompt_template) # Updated line

def chatbot_response(user_input):
    # Generate a response from the chatbot
    response = chat_chain.invoke({"user_input": user_input})
    return response

```

```python
user_input = input("You: ")
response = chatbot_response(user_input)
print("Chatbot:", response)
```
